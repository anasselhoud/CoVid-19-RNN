# **Problem 3**: 
The prime factors of 13195 are 5, 7, 13 and 29.

What is the largest prime factor of the number 600851475143 ?


--

**Hint:** 

1.  A prime number is a number greater than 1 and that cannot be formed by multiplying two smaller natural numbers but the number itself and 1. For example, 5 is a prime number because it can be devided only by 5 and 1.
2.  A prime factor is a factor that is a prime number. For example, 3 and 5 are prime factors of 15 because $3\times5=15$ and 3 and 5 are prime numbers.

<font color='red'>The code below would take too much time to be run for 600851475143, you can try and check it for 13195 first.</font>




```
#function to check if the number is a prime

def prime(a):

  if a==1:
    return False
  if a==2:
    return True

  for i in range(2,int(a/2)+1):
    if a%i==0:
      return False
  
  return True

def factors(n):
  Fact=[]
  k=1
  for i in range(2,int(n/2)+1):
    if n%i==0:
      Fact.append(i)
      
  return Fact

#Listing all factors of the number
Fact=factors(13195)

#sorting decreasingly the list
Fact.sort()

for i in Fact:
  if prime(i):
    k=i

print('The largest prime factor of the number above is',k)
```

# **Problem 4**: 
A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 × 99.

Find the largest palindrome made from the product of two 3-digit numbers.

--

**Hint:** 
We start first creating a function that return "True" when the number is palindrome. After that we map all the numbers between 99 and 999, calculate their product and check if the result is palindrome.

```
def palind(n):
  m=n
  r=0
  while m>0:
    u=m%10
    r=r*10+u
    m=m//10
  
  if (n==r):
    return True
  else:
    return False 
  
#Searching for the largest palinfrome made from 3-digitx3-digit
L=[]
for i in range(100,1000):
  for j in range(100,1000):
    N=i*j
    if palind(N):
      L.append(N)

L.sort(reverse = True)
print('The largest palindrome made from the product of two 3-digit numbers is',L[0])
```
